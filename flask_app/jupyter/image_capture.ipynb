```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
from shapely.geometry import Point

# Assuming gdf is your GeoDataFrame loaded from 'centroids_data.geojson'

for index, row in gdf.iterrows():
    # Wrap the geometry in a GeoDataFrame
    temp_gdf = gpd.GeoDataFrame(index=[0], crs=gdf.crs, geometry=[row.geometry])
    
    # Project the temporary GeoDataFrame to Web Mercator
    temp_gdf = temp_gdf.to_crs(epsg=3857)
    
    # Now temp_gdf is in Web Mercator projection, you can plot it
    fig, ax = plt.subplots(figsize=(8, 8))
    temp_gdf.plot(ax=ax, color='red', markersize=50)
    
    # Determine the extent to zoom into, using the buffer method on the geometry
    # Adjust buffer_size to your needs for zoom level
    buffer_size = 1000  # meters
    geom = temp_gdf.geometry.iloc[0]
    minx, miny, maxx, maxy = geom.buffer(buffer_size).bounds
    ax.set_xlim(minx, maxx)
    ax.set_ylim(miny, maxy)
    
    # Add basemap
    ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
    
    ax.set_axis_off()
    
    # Save the plot
    filename = f'../static/images/map_snapshot_{index}.png'
    plt.savefig(filename, dpi=72, bbox_inches='tight')
    plt.close()  # Close the plot to free memory


```
